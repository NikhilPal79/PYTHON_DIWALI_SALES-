```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
#import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status','unnamed1'], axis=1, inplace= True)
```

```python
#checking dropping executed 
df.info()
```

```python
##check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
#checking null value dropped 
pd.isnull(df).sum()
```

```python
# change data type 
df['Amount'] = df['Amount'].astype('int')
```

```python
# checking data type of amount changed
df.info()
```

```python
df.columns
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

```python



                                                            Exploratory Data Analysis



```

```python



                                                                      GENDER

```

```python
# plotting a bar chart for gender vs total amount

ax = sns.countplot(x= 'Gender',data= df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for Gender and its count

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender', y= 'Amount', data = sales_gen)
```

```python
From above graph we can see that most of the buyers are the female and having greater purchasing power than men 
```

```python


                                                                      AGE

```

```python
df.columns
```

```python
ax = sns.countplot(data= df, x= 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group', y= 'Amount', data = sales_age)
```

```python
From above graph we can clearly see that most of the buyers are of age group between 26-35yrs female 
```

```python



                                                                    STATE

```

```python
df.columns 
```

```python
ax = sns.countplot(data= df, x= 'State', hue = 'Orders')
for bars in ax.containers:
    ax.bar_label(bars)

sns.set(rc={'figure.figsize':(25,10)})
```

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data = sales_state, x = 'State', y ='Orders' )
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data = sales_state, x = 'State', y ='Amount' )
```

```python
From the above graph, we can see that most of the orders and total sales/amount are from Utter Pradesh, Maharashtra, Karnataka
```
